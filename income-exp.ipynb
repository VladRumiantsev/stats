```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
income = np.array([15000, 17500, 20000, 23000, 28800, 32400, 35000, 38650, 41000, 43500, 58000, 77000])
expenses = np.array([10000, 12000, 13500, 16600, 19500, 28000, 30000, 32600, 33800, 35000, 44100, 52000])
savings = np.array(income) - np.array(expenses)
plt.figure(figsize=(10, 6))
plt.plot(months, income, marker='o', color='red')
plt.title('monthly income', fontsize=14)
plt.xlabel('months', fontsize=14)
plt.ylabel('income ($)', fontsize=14)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(months, savings, color='blue')
plt.title('monthly savings', fontsize=14)
plt.xlabel('months', fontsize=14)
plt.ylabel('savings ($)', fontsize=14)
plt.tight_layout()
plt.show()

total_savings = np.sum(savings)
plt.figure(figsize=(8, 8))
plt.pie(savings, labels=months, autopct='%1.1f%%',
        colors=plt.cm.Paired(np.arange(len(months))))
plt.title('monthly savings in the total savings', fontsize=14)
plt.show()

quarter1 = income[0:3]
quarter2 = income[3:6]
quarter3 = income[6:9]
quarter4 = income[9:12]
average_quarters = {
    'Q1': stats.mean(quarter1),
    'Q2': stats.mean(quarter2),
    'Q3': stats.mean(quarter3),
    'Q4': stats.mean(quarter4)
}

for q, avg in average_quarters.items():
    print(f'average income for {q}: ${avg:.2f}')
```
